```python
import pandas as pd
# Define a dictionary containing Students data
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
       'Score 1' : [77,86,77,45],
       'Score 2' : [50,96,60,30]}
print("------before -------")
# Convert the dictionary into DataFrame
df = pd.DataFrame(data)

print(df)

print(df.sort_values(by='Score 1', ascending=True))

# sort by multiple columns
print(df.sort_values(by=['Score 1', "Name"], ascending=[True, True]))

```

```python
# example 2
import pandas as pd
df_cars = pd.read_json('cars.json')

print("------before -------")
# getting two column only
print(df_cars[['Car', 'quantity']])


print("------After Sorting -------")
# sorting base on ascending order in quantity 
df_cars.sort_values(by=['quantity'], ascending=[True])

# sorting base on quantity and car with respected value of ascending 
print(df_cars[['Car', 'quantity']].sort_values(by=['quantity','Car'], ascending=[False, True]))

print(df_cars.sort_values(by=['quantity','Car'], ascending=[False,True]))
```
